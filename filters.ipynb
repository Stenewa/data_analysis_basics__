```python
import pandas as pd
```

```python
people={"first":["jane","john","joseph"],"last":["doe","doe","carpt"],"email":["jane@gmail.com","john@gmail.com","jo@gmail.com"]}
people
```

```python
df=pd.DataFrame(people)
df
```

```python
filt=(df['last']=='doe')
filt
```

## Accessing data through filtered masks

### Eg.1

```python
df[filt]
```

### Eg.2

```python
df[df['last']=='doe']
```

### Eg.3

```python
df.loc[filt]
```

#### Specify col

```python
df.loc[filt,"email"]
```

### Using logical operators to filter data

```python
filt2=(df['last']=="doe")&(df['first']=="john")
```

```python
df.loc[filt2,"email"]
```

```python
filt3=(df["first"]=="john")|(df["last"]=="doe")
```

```python
df.loc[filt3,"first"]
```

### To negate

```python
df.loc[~filt3,"first"]
```
