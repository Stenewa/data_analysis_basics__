```python
this_dic={'first':['Ann','Leo','Mac','Pod',np.nan,np.nan,'NA'],
         'Last':['Inn','Lee','May','Pea',np.nan,np.nan,'Missing'],
         'Email':['Aninn@gmail.com','Lele@gmail.com','Mama@gmail.com',None,None,np.nan,'Missing'],
         'Age':['33','34','25','28',None,None,'Missing']}
this_dic

```

```python
diDf=pd.DataFrame(this_dic)

#replace custom missing values with nan
diDf.replace('Missing',np.nan,inplace=True)
diDf.replace('NA',np.nan,inplace=True)
diDf
```

```python
diDf.dropna()
```

```python
 #defaukt dropna() 
diDf.dropna(axis='index',how='any')
```

```python
diDf.dropna(axis='index',how='all')
```

```python
diDf.dropna(axis='index',how='any',subset=['Email'])
```

```python
diDf.dropna(axis='index',subset=['Email','Last'])
```

```python
diDf.dropna(axis='index',how='all',subset=['Email','Last'])
```
