```python
resgp=this_grp['Country'].value_counts()
resgp
```

```python
pygp=this_grp['LanguageHaveWorkedWith'].apply(lambda x:x.str.contains('Python').sum())
pygp
```

```python
serDf=pd.concat([resgp,pygp],axis='columns',sort=False)
serDf
```

```python
serDf.rename(columns={'count':'No.ofRespondents','LanguageHaveWorkedWith':'No.KnowPy'},inplace=True)
serDf
```

```python
serDf
```

```python
serDf['PctPy']=(serDf['No.KnowPy']/serDf['No.ofRespondents'])*100
serDf
```

```python
serDf.sort_values(by='PctPy',ascending=False,inplace=True)
serDf.head(50)
```
