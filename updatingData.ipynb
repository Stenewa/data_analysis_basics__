```python
import pandas as pd
```

```python
people={"first":["jane","john","joseph"],"last":["doe","doe","carpt"],"email":["jane@gmail.com","john@gmail.com","jo@gmail.com"]}
people
```

```python
peopleDf=pd.DataFrame(people)
peopleDf
```

```python
peopleDf.columns
```

### Updating column names

```python
peopleDf.columns=['fname','lname','email']
peopleDf
```

```python
peopleDf.columns=[x.upper() for x in peopleDf.columns]
peopleDf
```

```python
peopleDf.columns=[x.lower() for x in peopleDf.columns]
peopleDf
```

```python
peopleDf.rename(columns={"fname":"first","lname":"last"}, inplace=True)
peopleDf
```

### Changing Row data

```python
peopleDf.loc[2]
```

```python
peopleDf
```

```python
peopleDf.loc[0,["first","email"]]=["jaine","jaine@gmail.com"]
peopleDf
```

```python
peopleDf.loc[0,"first"]="jayne"
peopleDf.loc[0,'last']='doe'
peopleDf
```

```python
peopleDf.at[1,'first']="jon"
peopleDf
```

```python
pplFilt=peopleDf["last"]=="Smith"
peopleDf.loc[pplFilt,"last"]="SMYTHE"
peopleDf
```
