```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import csv
from collections import Counter
```

```python
with open('data.csv') as this_file:
    thiscsv= csv.DictReader(this_file)
    language_count=Counter()
    for row in thiscsv:
        language_count.update(row['LanguagesWorkedWith'].split(';'))
language_count.most_common(15)
```

```python
language=[]
popularity=[]
for item in language_count.most_common(15):
    language.append(item[0])
    popularity.append(item[1])
```

```python
language.reverse()
popularity.reverse()
plt.barh(language,popularity)
plt.tight_layout
plt.title('Popular Languages')
#plt.ylabel('Languages')
plt.xlabel('No. of People who use')
```
