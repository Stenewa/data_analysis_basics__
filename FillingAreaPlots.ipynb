```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import csv
from collections import Counter
```

```python
fill_data=pd.read_csv('fill_data.csv')
py_sal=np.array(fill_data['Python'])
js_sal=np.array(fill_data['JavaScript'])
all_sal=np.array(fill_data['All_Devs'])
age=fill_data['Age']
```

```python
plt.plot(dev_x,dev_y,color='k',linestyle='--',label='AllDevs')
plt.plot(dev_x,py_dev_y,color='b',linewidth=1,label='PythonDevs')
plt.plot(dev_x,js_dev_y,label='JsDevs')
plt.title('Medain Dev Salary(USD)')
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.legend(['AllDevs','PythonDevs','Js Devs'])
plt.style.use('seaborn-pastel')
plt.fill_between(dev_x,py_dev_y,med,where=(py_dev_y>med),interpolate=True,alpha=0.1)
plt.show()
```
